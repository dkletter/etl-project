```python
json_business = 'data/yelp_academic_dataset_business.json'
business_df = pd.read_json(json_business, lines=True)
business_df.head()
```

```python
clean_business_df = business_df[['name', 'address', 'city', 'state', 'postal_code', 'stars', 'review_count', 'categories']].copy().dropna()
clean_business_df.head()
```

```python
keywords = ["Restaurants", "Coffee & Tea"]
clean_business_df = clean_business_df[clean_business_df['categories'].str.contains('|'.join(keywords))]
clean_business_df.head()
```
